```python
plt.figure(figsize=(6, 6))
sns.barplot(data=data, y=data['state'].sort_values(ascending=True), x='PM10', palette='viridis')
```

```python
plt.figure(figsize=(8,16));
sns.jointplot(x='state', y='status', data=data, kind='scatter',space=0.2,palette="coolwarm");
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(34,8));
sns.swarmplot(x="state", y="status", data=data)
plt.xticks(rotation = 45)
```
