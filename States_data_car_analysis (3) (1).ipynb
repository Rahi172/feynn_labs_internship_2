```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
```

```python
data=pd.read_csv("states_data_car.csv")
```

```python
plt.rcParams['figure.figsize'] = (11, 13)
data.hist()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=data, y=data['state'].sort_values(ascending=True), x='petrol', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Petrol Prices', family='serif', fontsize=14, labelpad=10)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title(label='Petrol Price at different states', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=data, y=data['state'].sort_values(ascending=True), x='diesel', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Diesel Prices', family='serif', fontsize=14, labelpad=10)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.title(label='Diesel Price at different states', weight=200, family='serif', size=15, pad=12)
plt.show()
```
